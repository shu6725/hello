```python
def is_integer_matrix(matrix):
    a = True
    for i in range(0, 3):
        for j in range(0, 3):
            if not matrix[i][j].is_integer():
                a = False
                break
    return a
```

```python
# reduced_HNF[1]について
H = reduced_HNF[1]
H_inv = np.linalg.inv(H)

for i in range(len(parent_sym["rotations"])):
    check = H_inv.dot(parent_sym["rotations"][i].dot(H))
    if is_integer_matrix(check):
        print(check)
```
