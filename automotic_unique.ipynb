```python
def trans_unique(omomi4, trans_tikans):
    omomi4_neo = deepcopy(omomi4)
    lis = set()
    superperiodic = set()
    for i in range(len(omomi4)):#through all candidate
        if omomi4[i] in omomi4_neo:#kouho ga mada 生き残ってるかチェック
            for j in range(1, len(trans_tikans)):#置換操作について回す　
                d = dict()
                sta = ""
                for k in range(len(trans_tikans[0])):
                    d[trans_tikans[j][k]] = omomi4[i][k]#str辞書の作成
                for r  in range(len(trans_tikans[0])):#staの作成 sta is made from tikan[j]
                    sta += d[r]
                if int(sta) == int(omomi4[i]):
                    superperiodic.add(sta)
                    lis.add(omomi4[i])
                else:
                    if sta in omomi4_neo:
                        omomi4_neo.remove(sta)
                        lis.add(omomi4[i]) 
    lis = lis - superperiodic
    lis = list(lis)
    return lis
```

```python
def sucell_unique(rutile, HNF, index):
    parent_lattice = generate_superlattice.get_superlattice(rutile, HNF, index)
    o_sublattice = generate_superlattice.get_o_superlattice(rutile, HNF, index)    
    parent_lattice_jun = permutation.generate_per(parent_lattice, o_sublattice)
    ds_list = list()
    for oxy_number in range(index*4):
        can = permutation.make_candidate(o_sublattice, oxy_number)
        can = trans_unique(can, trans_tikans)
        auncko = permutation.unique(parent_lattice_jun, can)
        ds_list.extend(auncko)
    return ds_list
```
