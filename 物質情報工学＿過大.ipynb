```python
import numpy
import matplotlib.pyplot as plt

a = numpy.random.rand(5000)
b = numpy.random.rand(5000)
c = numpy.random.rand(5000)
y = a + b + c
y = numpy.round(y, 2)
plt.hist(y, bins=100)
plt.show()
```
